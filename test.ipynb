```python
import osmnx as ox
from IPython.display import HTML

ox.settings.use_cache = False
```

```python
place = "Piedmont, California, USA"
G = ox.graph_from_place(place, network_type="drive")
Gp = ox.project_graph(G)

orig = list(G)[0]
dest = list(G)[120]
route = ox.shortest_path(G, orig, dest, weight="length")
fig, ax = ox.plot_graph_route(G, route, route_color="y", route_linewidth=6, node_size=0)
```

```python
from osmnx.plot import animate_graph_route

fig, ax, anim = animate_graph_route(G, route, route_color="r", route_linewidth=5, node_size=0, route_alpha=1, speed=8)
```
